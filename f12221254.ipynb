```python
d.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
d['duration'].fillna(0, inplace=True)
d['class'].fillna('Unknown', inplace=True)
d['airline'].fillna('Unknown', inplace=True)
```

```python
d.info()
```

```python
d
```

```python
d['departure_time'] = d['departure_time'].str.replace('Early_Morning', 'Morning')
d['arrival_time'] = d['arrival_time'].str.replace('Early_Morning', 'Morning')
```

```python
d['arrival_time'] = d['arrival_time'].str.replace('Late_Night', 'Night')
d['departure_time'] = d['departure_time'].str.replace('Late_Night', 'Night')
```

```python

d['airline'] = d['airline'].str.replace('Air_India', 'Air India')

d['airline'] = d['airline'].str.replace('AirAsia', 'Air Asia')
d['airline'] = d['airline'].str.replace('GO_FIRST', 'Go First')
```

```python
# Filtering out the duplicate entries
d = d[~d.duplicated()]
d[d.duplicated()]
```

```python
# Changing the data type
d['price'] = pd.to_numeric(d['price'], errors = 'coerce')
```

# Max and Min prices in the given data

```python
# Get the maximum price
max_price = d['price'].max()
print(f"Maximum price: {max_price}")

# Get the minimum price
min_price = d['price'].min()
print(f"Minimum price: {min_price}")
```

```python
bins = [0, 5000, 10000, 15000, np.inf]
labels = ['Budget-friendly', 'Mid-range', 'Premium', 'Luxury']
d['price_bins'] = pd.cut(d['price'], bins=bins, labels=labels)
```

```python
d
```

```python
bins = [0, 60, 120, 180, 240, np.inf]
labels = ['Very Short (<10min)', 'Short (10-30min)', 'Medium (30-60min)', 'Long (1-2hrs)', 'Very Long (>2hrs)']
d['duration_bins'] = pd.cut(d['duration']*60, bins=bins, labels=labels)
```
